```python
f = open("common_passwords.txt", "r")
contents = f.readlines()
```

```python
char_count_frequency = {}
total_count = 0

for word in contents:
    word = word[:-1]
    for i in range(len(word)):
        if word[i] not in char_count_frequency:
            char_count_frequency[word[i]] = 1
        else:
            char_count_frequency[word[i]] += 1
        total_count += 1    

print(total_count)
char_count_frequency
```

```python
print(total_count)

cur_count = 0
for char, count in char_count_frequency.items():
    cur_count += count
    print(char)
    print('%d' % (cur_count))
```

```python
file = open("char_freq_distribution.txt", "w")

file.write('%d' % (total_count) + "\n")

cur_count = 0
for char, count in char_count_frequency.items():
    cur_count += count
    file.write(char + "\n")
    file.write('%d' % (cur_count) + "\n")

file.close()
```
